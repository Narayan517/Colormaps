```python
#PLOTTING WORLD MAP
!pip install geopandas

import geopandas as gpd
import matplotlib.pyplot as plt

# Load the world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the map
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color='lightblue', edgecolor='black')
ax.set_title('World Map')
plt.axis('off')
plt.show()
```

```python
#COLORING CONTINENTS
# Load the world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the map with colors based on continent
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, column='continent', legend=True,
            legend_kwds={'loc': 'upper left'}, cmap='tab10')
ax.set_title('World Map with Coloring by Continent')
plt.axis('off')
plt.show()
```

```python
#HIGHLIGHT INDIA

# Load the world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the map with India highlighted in red
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color='lightblue', edgecolor='black')
Australia = world[world.name == 'Australia']
Australia.plot(ax=ax, color='red', edgecolor='black', alpha=0.8)  # Highlight India in red
ax.set_title('World Map with India Highlighted')
plt.axis('off')
plt.show()
```

```python
#MAP OF ASIA
# Load the world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Filter countries in Asia
asia = world[world['continent'] == 'Asia']

# Plot the map with country labels and different colors
fig, ax = plt.subplots(figsize=(12, 8))
asia.plot(ax=ax, color='lightgray', edgecolor='black', cmap='tab10')

for x, y, label in zip(asia.geometry.centroid.x, asia.geometry.centroid.y, asia['name']):
    ax.text(x, y, label, fontsize=8, ha='center', va='center')

ax.set_title('Map of Asia with Country Labels')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.axis('off')
plt.show()
```
